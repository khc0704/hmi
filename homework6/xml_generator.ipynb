```python
import shutil
import random
import string
from skimage import io
```

```python
headstr="""\
<annotation>
	<folder>Images</folder>
	<filename>%06d.jpg</filename>
	<path>%s</path>
	<source>
		<database>Unknown</database>
	</source>
	<size>
		<width>%d</width>
		<height>%d</height>
		<depth>%d</depth>
	</size>
	<segmented>0</segmented>
"""

objstr="""\
	<object>
		<name>%s</name>
		<pose>Unspecified</pose>
		<truncated>0</truncated>
		<difficult>0</difficult>
		<bndbox>
			<xmin>%d</xmin>
			<ymin>%d</ymin>
			<xmax>%d</xmax>
			<ymax>%d</ymax>
		</bndbox>
	</object>
"""

tailstr="""\
</annotation>
"""

Annotations_DIR = os.path.join(DATASETS_DIR,"Annotations")
ImageNames_DIR = os.path.join(DATASETS_DIR,"ImageNames/Main")
Images_DIR = os.path.join(DATASETS_DIR,"Images")

def write_xml(idx,head,objs,tail):
  filename = os.path.join(Annotations_DIR,("%06d.xml" % (idx)))
  f = open (filename,"w")
  f.write(head)
  for obj in objs:
    f.write(objstr % ('face',obj[0],obj[1],obj[0]+obj[2],obj[1]+obj[3]))
  f.write(tail)
  f.close()

def generate_datasets(idx,datatype):

  f_name = open(os.path.join(ImageNames_DIR,(datatype+'.txt')),'a')
  f_anno = open((os.path.join(DATASETS_DIR,('wider_face_split/wider_face_'+datatype+'_bbx_gt.txt'))),'r')

  while True:
    filename = f_anno.readline().strip('\n')
    file_path = os.path.join(DATASETS_DIR,('WIDER_'+datatype+'/images/'+filename))
    new_file_path = os.path.join(Images_DIR,('%06d.jpg' % (idx)))
    if not filename:
      break
    
    img = io.imread(file_path)
    head = headstr % (idx,new_file_path,img.shape[1],img.shape[0],img.shape[2])
    objs_num = f_anno.readline().strip('\n')
    objs = []

    if objs_num == '0':
      object_info = f_anno.readline()
      continue
    
    for i in range(int(objs_num)):
      obj_info = f_anno.readline().strip(' \n').split(' ')
      obj = [int(info) for info in obj_info]

      if obj[7] == 0:
        objs.append(obj)
    
    write_xml(idx,head,objs,tailstr)
    shutil.copyfile((file_path),(new_file_path))
    f_name.write('%06d\n' % (idx))
    if idx%100 == 0:
      print(idx)
      print(datatype)
    idx += 1
  f_name.close()
  f_anno.close()
  return idx


def create_dir():
  single_DIRS = [(Annotations_DIR),(Images_DIR)]
  for DIR in single_DIRS:
    if shutil.os.path.exists(DIR):
      shutil.rmtree(DIR)
    shutil.os.mkdir(DIR)

  if shutil.os.path.exists(ImageNames_DIR):
    shutil.rmtree('/'.join(ImageNames_DIR.split('/')[:-1]))
  shutil.os.makedirs(ImageNames_DIR)

```

```python
create_dir()
idx = 1
idx = generate_datasets(idx,'train')
idx = generate_datasets(idx,'val')
print('Done')
```
