```python
import os
ROOT_DIR = "/content/gdrive/yolov4"
#ROOT_DIR = "/content/gdrive/MyDrive/Colab Notebooks/HMI/final_project"
DATASETS_DIR = os.path.join(ROOT_DIR,"datasets")
%cd $ROOT_DIR
%cd datasets
```

```python
import xml.etree.ElementTree as ET
import shutil
```

```python
Annotations_DIR = os.path.join(DATASETS_DIR,"Annotations")
ImageNames_DIR = os.path.join(DATASETS_DIR,"ImageNames/Main")
Images_DIR = os.path.join(DATASETS_DIR,"Images")
Labels_DIR = os.path.join(DATASETS_DIR,"labels")

cls = ["face"]

def convert(info):
  dw = 1. / info["width"]
  dh = 1. / info["height"]
  x = (info["xmin"]+info["xmax"]) / 2.0
  y = (info["ymin"]+info["ymax"]) / 2.0
  w = info["xmax"] - info["xmin"]
  h = info["ymax"] - info["ymin"]
  x = x * dw
  w = w * dw
  y = y * dh
  h = h * dh
  return (x,y,w,h)

def convert_yolo_data(idx):
  input_file = open(os.path.join(Annotations_DIR,('%s.xml' % (idx))))
  output_file = open(os.path.join(Images_DIR,('%s.txt' % (idx))),'w')

  tree = ET.parse(input_file)
  root = tree.getroot()
  size = root.find('size')
  w = int(size.find('width').text)
  h = int(size.find('height').text)
  obj_num = 0
  for obj in root.iter('object'):
    diff  = obj.find('difficult').text
    name = obj.find('name').text
    if name not in cls or int(diff) == 1:
      continue
    box = obj.find('bndbox')
    info = {"width":w,"height":h,"xmin":float(box.find('xmin').text),"xmax":float(box.find('xmax').text),
            "ymin":float(box.find('ymin').text),"ymax":float(box.find('ymax').text)}
    new_info = convert(info)
    error_obj = False
    for index,new_info_check in enumerate(new_info):
      if new_info_check == 0 and index >=2 :
        error_obj = True
        break
    if error_obj == False:
      message = str(cls.index(name))+" "+" ".join([str(tmp) for tmp in new_info])+"\n"
      output_file.write(message)
      obj_num += 1
  input_file.close()
  output_file.close()
  if obj_num == 0:
    return True
  else:
    return False

def create_dir():
  if shutil.os.path.exists(Labels_DIR):
    shutil.rmtree(Labels_DIR)
  shutil.os.mkdir(Labels_DIR)
  for datatype in ('train','val'):
    path = os.path.join(DATASETS_DIR,('%s.txt' % datatype))
    if shutil.os.path.exists(path):
      os.remove(path)
    
```

```python
create_dir()
for datatype in ('train','val'):
  image_name_file = open(os.path.join(ImageNames_DIR,('%s.txt' % datatype)))
  idxs = image_name_file.read().strip().split()
  data_path_file = open(os.path.join(DATASETS_DIR,('%s.txt' % datatype)),'w')
  for idx in idxs:
    err = convert_yolo_data(idx)
    if err:
      print(idx,' only contain error label')
      continue
    else:
      data_path = os.path.join(Images_DIR,('%s.jpg' % idx))+'\n'
      data_path_file.write(data_path.replace("\\",'/'))
    if int(idx) % 100 ==0:
      print(idx)
  image_name_file.close()
  data_path_file.close()
```
